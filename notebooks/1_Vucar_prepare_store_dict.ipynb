```python
from collections import defaultdict, Counter
import json
import pandas as pd

data_dict = defaultdict(lambda: defaultdict(lambda: defaultdict(lambda: {"origin": [], "type": [], "seats": [], "gearbox": [], "fuel": [], "color": []})))

for _, row in df_copy.iterrows():
    manufacture_date = row["manufacture_date"]
    brand = row["brand"]
    model = row["model"]

    if not pd.isna(row["origin"]):
        data_dict[manufacture_date][brand][model]["origin"].append(row["origin"])
    if not pd.isna(row["type"]):
        data_dict[manufacture_date][brand][model]["type"].append(row["type"])
    if not pd.isna(row["seats"]):
        data_dict[manufacture_date][brand][model]["seats"].append(row["seats"])
    if not pd.isna(row["gearbox"]):
        data_dict[manufacture_date][brand][model]["gearbox"].append(row["gearbox"])
    if not pd.isna(row["fuel"]):
        data_dict[manufacture_date][brand][model]["fuel"].append(row["fuel"])
    if not pd.isna(row["color"]):
        data_dict[manufacture_date][brand][model]["color"].append(row["color"])
```

```python
with open('data_dict.json', 'w', encoding='utf-8') as f:
    json.dump(data_dict, f, ensure_ascii=False, indent=4)
```

```python
def get_most_frequent_values(nested_dict):
    final_data = {}
    for m_date, brands in nested_dict.items():
        final_data[m_date] = {}
        for brand, models in brands.items():
            final_data[m_date][brand] = {}
            for model, attributes in models.items():
                final_data[m_date][brand][model] = {}
                for attr, values_list in attributes.items():
                    # print(values_list)
                    if values_list:
                        most_common = Counter(values_list).most_common()
                        max_count = most_common[0][1]  # Get the highest count

                        candidates = [val for val, count in most_common if count == max_count]
                        final_data[m_date][brand][model][attr] = candidates[0]
                    else:
                        final_data[m_date][brand][model][attr] = fetch_from_openAI_api(m_date, brand, model, attr)
                        
    return final_data
```

```python
def fetch_from_openAI_api(m_date, brand, model, attr):
    unique_values = df_copy[attr].unique().tolist()  

    prompt = (
        f"Based on the following information:\n"
        f"Manufacture Date: {m_date}\n"
        f"Brand: {brand}\n"
        f"Model: {model}\n"
        f"Please predict the value for the attribute '{attr}' from the following options:\n"
        f"{', '.join([str(value) for value in unique_values])}.\n"
        f"Return only the most likely single value without no explanation."
    )

    try:
        response = openai.ChatCompletion.create( 
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "You are a helpful assistant."},
                {"role": "user", "content": prompt}
            ],
            max_tokens=50,  # Limit the response length
            temperature=0.3,  # Low randomness for more consistent results
            n=1,  # Get only one result
        )
        predicted_value = response["choices"][0]["message"]["content"].strip()  # Extract prediction
        if predicted_value not in unique_values:
            # Fallback if the response is invalid
            return unique_values[0] if unique_values else "Unknown"
        print(predicted_value)
        return predicted_value
    except Exception as e:
        print(f"Error in API call: {e}")
        return "Unknown"
```

```python
filled_data_dict = get_most_frequent_values(data_dict)

sorted_filled_data_dict = dict(sorted(filled_data_dict.items(), key=lambda x: x[0], reverse=True))
```
