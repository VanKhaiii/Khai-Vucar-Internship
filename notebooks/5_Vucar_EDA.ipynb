## 1. How does mileage relate to car price? Is there a clear negative correlation?

```python
columns_of_interest = ['price', 'mileage_v2']

df[columns_of_interest].describe()
```

**Comments:**
1. Cả `price`và `mileage_v2` đều có std (độ lệch chuẩn khá lớn) (~684,17 triệu cho `price` và ~92.295 km cho `mileage_v2`), cho thấy sự giao động và sự không ổn định cho các giá trị ở hai cột thuộc tính này.
2. Cả `price` và `mileage_v2` đều có max và min khá lệch nhau, cho thấy dữ liệu này rất đa dạng và không được ổn định.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='mileage_v2', y='price', data=df)
plt.title('Mối quan hệ giữa mileage và price')
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.show()
```

**Comments:**
1. Xu hướng tổng quan: Biểu đồ thể hiện một mối quan hệ giảm dần giữa `mileage_v2` và `price`. Cụ thể:
     - Khi quãng đường đi (`mileage_v2`) tăng lên, giá (`price`) thường có xu hướng giảm.
     - Đây là một xu hướng hợp lý, vì xe có quãng đường đi càng nhiều thì thường có giá bán thấp hơn.

2. Ngoại lệ (Outliers):
   - Có một số điểm nằm ở phần `price` rất cao (lên đến hơn 10 tỷ), trong khi `mileage_v2` rất thấp. Những điểm này có thể là: Dữ liệu của các loại xe cao cấp, xe mới tinh chưa qua sử dụng.
   
3. Phân bố dữ liệu:
   - Phần lớn các điểm dữ liệu tập trung ở quãng đường từ **0 đến 200.000 km** và giá từ **0 đến 1 tỷ**.
   - Các xe có quãng đường lớn hơn **500.000 km** rất hiếm, nhưng vẫn tồn tại một vài điểm dữ liệu với giá thấp.

```python
correlation = df['mileage_v2'].corr(df['price'])
print(f"\nHệ số tương quan giữa mileage và price: {correlation:.2f}")
```

```python
if correlation < -0.5:
    print("Có mối quan hệ nghịch biến mạnh giữa số dặm và giá xe.")
elif -0.5 <= correlation < -0.3:
    print("Có mối quan hệ nghịch biến trung bình giữa số dặm và giá xe.")
elif -0.3 <= correlation < 0:
    print("Có mối quan hệ nghịch biến yếu giữa số dặm và giá xe.")
else:
    print("Không có mối quan hệ nghịch biến rõ ràng giữa số dặm và giá xe.")
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='fuel', y='price', data=df)
plt.title('Phân phối giá xe theo loại nhiên liệu')
plt.xlabel('Loại nhiên liệu')
plt.ylabel('Giá xe')
plt.xticks(rotation=45)
plt.show()
```

**Comments:** Biểu đồ boxplot cho thấy sự phân phối giá xe của 4 loại nhiên liệu: `petrol`, `oil`, `electric`, và `hybrid`.
1. `Petrol`:
 - Phân phối giá có sự trải rộng và nhiều giá trị ngoại lệ (outliers) với mức giá cao vượt trội (trên 1 tỷ đến hơn 10 tỷ).
 - Phần lớn các xe xăng có giá dao động trong khoảng từ vài trăm triệu đến dưới 1 tỷ.

2. `Oil`:
 - Phân phối giá của xe dầu thấp hơn so với xe xăng.
 - Có một vài điểm ngoại lệ nhưng ít hơn đáng kể so với xe xăng.

3. `Electric`:
 - Phân phối giá xe điện tương đối thấp.
 - Số lượng outliers cũng ít hơn so với xe xăng và dầu.

4. `Hybrid`:
 - Giá xe hybrid khá ổn định, không nhiều biến động lớn.

```python
mean_prices_by_fuel = df.groupby('fuel')['price'].mean().sort_values(ascending=False)
print("\nGiá trung bình theo loại nhiên liệu:\n", mean_prices_by_fuel)
```

1. Đầu tiên, em tính giá trị chêch lệch của giá trị cao nhất và thấp nhất trong nhiên liệu.
2. Sau đó kiểm tra xem sự chênh lệch này có lớn hơn 20% mức trung bình chung hay không? Nếu có, điều này cho thấy loại nhiên liệu góp phần lớn trong việc định giá.

```python
max_price_diff = mean_prices_by_fuel.max() - mean_prices_by_fuel.min()
if max_price_diff > (0.2 * mean_prices_by_fuel.mean()):
    print("Loại nhiên liệu có ảnh hưởng đáng kể đến giá xe.")
else:
    print("Loại nhiên liệu không có ảnh hưởng đáng kể đến giá xe.")
```

## 3. How to compare two cars in the same segment?

1. Vì không thể biết được các phân khúc xe sẽ được quy ước như thế nào? Nên em quyết định dựa vào giá, chia cột `price` thành 8 đoạn, sau đó bắt đầu đánh phân khúc cho các loại xe này từ A -> H.

```python
df['segment'] = pd.qcut(df['price'], q=8, labels=list("HGFEDCBA"))
df.head(10)
```

```python
def compare_cars(car1_id, car2_id):
    try:
        car1 = df[df['id'] == car1_id].iloc[0]
        car2 = df[df['id'] == car2_id].iloc[0]

        if car1['segment'] != car2['segment']:
            print("Hai xe không thuộc cùng phân khúc. Hãy chọn lại.")
            return

        comparison = pd.DataFrame({
            "Thuộc tính": ["Hãng xe", "Mẫu xe", "Năm sản xuất", "Giá", "Số dặm", "Loại nhiên liệu", "Phân khúc"],
            "Xe 1": [car1['brand'], car1['model'], car1['manufacture_date'], car1['price'], car1['mileage_v2'], car1['fuel'], car1['segment']],
            "Xe 2": [car2['brand'], car2['model'], car2['manufacture_date'], car2['price'], car2['mileage_v2'], car2['fuel'], car2['segment']]
        })

        print("\nSo sánh chi tiết hai xe:\n")
        print(comparison)

        # Vẽ biểu đồ để minh họa sự khác biệt về giá và mileage
        plt.figure(figsize=(10, 5))
        plt.bar(["Car 1 - Giá", "Car 2 - Giá"], [car1['price'], car2['price']], color=['blue', 'green'])
        plt.title("So sánh giá giữa hai xe")
        plt.show()

        plt.figure(figsize=(10, 5))
        plt.bar(["Car 1 - Mileage", "Car 2 - Mileage"], [car1['mileage_v2'], car2['mileage_v2']], color=['orange', 'red'])
        plt.title("So sánh số dặm giữa hai xe")
        plt.show()

    except KeyError:
        print("ID xe không hợp lệ. Vui lòng kiểm tra lại.")
```
